```python
#importing libraries
import math
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#Loading data set
df = pd.read_excel('Copper Daily Prices.xlsx').reset_index(drop = True)
df.head()
```

```python
df.dtypes
```

```python
# Data Formatting
df['Date_1'] = df['date'].apply(pd.to_datetime)
df_formatted = df[['Date_1','LME Copper Cash-Settlement']]
df_formatted = df_formatted.rename(columns={'Date_1': 'Date', 'LME Copper Cash-Settlement': 'Copper_Price'})
df_formatted = df_formatted.sort_values(by = ['Date']).reset_index(drop = True)
df_formatted.head()
```

```python
df_formatted.describe()
```

```python
#EDA and Data Visualization
plt.figure(figsize=(16,8))
plt.title('Price History of Copper')
plt.plot(df_formatted['Date'], df_formatted['Copper_Price'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('Price (in Rs.)',fontsize=18)
plt.show()
```

### Preparing Training Data

```python
#Create a new data frame with only the 'Close column'
data = df_formatted.filter(['Copper_Price'])

#Convert the dataframe to numpy array
dataset = data.values

#Get the number of rows to train the model on (using 90% data for the training)
training_data_len = math.ceil(len(dataset) * 0.9 )

training_data_len
```

```python
#Feature Scaling (Normalisation as using the neural network)
#to ensure that gradient descent converges more smoothly and quickly, errors are much smaller 
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
#Only take Training data
scaled_data
```

```python
len(scaled_data)
```

#### Preparing Training data for time series with Lookback method

```python
#Create the training data set
#Create the scaled training set
train_data = scaled_data[0:training_data_len , :]

# generate the input and output sequences
n_lookback = 270  # length of input sequences (lookback period)
n_forecast = 90  # length of output sequences (forecast period)

#Lookback period should be multiple of forecast days

#Split the data into x_train and _train data sets
x_train = []
y_train = []


#iterate from lookback index and till size - forecast days so that we get shape of X & Y consistent
for i in range(n_lookback, len(train_data) - n_forecast + 1):
    #X_train is list of list, list will be of 180 sequencial element, and element is "train_data[1:181,0]"
  x_train.append(train_data[i-n_lookback:i,0])
    #y_train is list of element, element will be train_data[181,181+90]"
  y_train.append(train_data[i:i+n_forecast,0])

# For the sequencial 120 training element, the test element will be 121th element, as it is time-series data
#Below is the sample of the training data

#Convetring list to numpy array
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
y_train.shape
```

```python
# Reshaping the array to 3 dimentional to feed to the model 

x_train_1 = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
# Input shape should be same as data points in X
input_shape= (x_train_1.shape[1], 1)
input_shape
```

```python
#Build the LSTM model
model = Sequential()
#A classical LSTM cell already contains quite a few non-linearities: 
#three sigmoid functions and one hyperbolic tangent (tanh) function
model.add(LSTM(180, return_sequences=True, input_shape= (x_train_1.shape[1], 1)))
model.add(LSTM(150, return_sequences=False))

model.add(Dense(120))
model.add(Dense(100))
model.add(Dense(90))
```

```python
print(model.summary())
```

```python
# Compile the model
# the Adam optimizer is straightforward to implement, has faster running time, 
# low memory requirements, and requires less tuning than any other optimization algorithm.
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(x_train_1, y_train, batch_size=30, epochs=3)
```

```python
x_train_1.shape
```

## Validate the Model on 10% of the data
##### Creating test Data

```python
#Validate the Model
test_data = scaled_data[training_data_len - n_lookback: , :]


#Create the data sets x_test and y_test
x_test = []
y_test = []

for i in range(n_lookback, len(test_data) - n_forecast + 1):
  x_test.append(test_data[i-n_lookback:i, 0])
  y_test.append(test_data[i:i+n_forecast,0])

x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
x_test.shape
```

```python
x_test_1 = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
#Model prediction on Test data
Y_test = model.predict(x_test_1)
Y_test = scaler.inverse_transform(Y_test)
```

```python
Val_score = model.evaluate(x_test_1, y_test, batch_size=30)
print(' Validation Loss :', Val_score)
```

```python
#Extracting Y validation values predicted by LSTM model
Y_test_1 = []
for i in range(len(Y_test)):
    if i < len(Y_test)-1:
        
        Y_test_1.append(Y_test[i][0])
        
    else:
        for j in Y_test[i]:
              Y_test_1.append(j)
                
Y_test_1 = np.array(Y_test_1)
Y_test_1 = np.reshape(Y_test_1, (Y_test_1.shape[0], 1))
```

```python
Y_test_1[-10:]
```

```python
#Taking Last n_lookback Data points to Predict the n_forecast points in future
X_Predict = scaled_data[-n_lookback:]
X_Predict = X_Predict.reshape(1, n_lookback, 1)

#LSTM Model to predict the price for n_forecast days
Y_Predict = model.predict(X_Predict).reshape(-1, 1)
Y_Predict = scaler.inverse_transform(Y_Predict)
```

#### Actual and Validated Values

```python
# Past Data
df_past = df_formatted
df_past.rename(columns={'Copper_Price': 'Actual'}, inplace=True)
df_past['Forecast'] = np.nan
#For last len(Y_test_1) values in dataset, adding values predicted by Model
df_past['Forecast'].iloc[-len(Y_test_1):] = Y_test_1.flatten()

df_past.tail(10)
```

### 90 days Forecasted values

```python
df_future = pd.DataFrame(columns=['Date', 'Actual', 'Forecast'])
df_future['Date'] = pd.date_range(start=df_past['Date'].iloc[-1] + pd.Timedelta(days=1), periods=n_forecast)
df_future['Forecast'] = Y_Predict.flatten()
df_future['Actual'] = np.nan
df_future
```

```python
len(df_future)
```

```python
#Appending Actual and Forecasted values
results = df_past.append(df_future).set_index('Date')
results.tail()
```

### Plots for evaluation

```python
# plot the results
plt.figure(figsize=(30,15))
plt.title('Forecasting Copper Prices')
plt.xlabel('Date', fontsize=35)
plt.ylabel('LME Copper price in INR', fontsize=35)
plt.plot(results)
plt.legend(['Actual', 'Predicted'], loc='lower right')
plt.show()
```
