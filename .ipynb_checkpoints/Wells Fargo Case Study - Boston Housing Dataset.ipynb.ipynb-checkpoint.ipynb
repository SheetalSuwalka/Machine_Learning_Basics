#### 4f. Divide the Data into Train/Test Set

```python
##################
# YOUR CODE HERE
#################
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df_filter, df_Y, test_size =  0.2, random_state = 35)
```

```python
Y_train.head()
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y_train = le.fit_transform(Y_train)
```

```python
Y_train
```

#### 4g. Train algorithm(s) of your choice to predict the housing price (include HyperParameter Tuning if required)

```python
##################
# YOUR CODE HERE
#################
from xgboost import XGBClassifier
clf = XGBClassifier(objective='reg:squarederror')
clf.fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred = le.inverse_transform(y_pred)
y_pred
```

```python
ytrain_pred = clf.predict(X_train)
ytrain_pred = le.inverse_transform(ytrain_pred)
ytrain_pred
```

```python
from sklearn.metrics import mean_squared_error
mse_test = mean_squared_error(Y_test, y_pred)
mse_test
```

```python
mse_train = mean_squared_error(Y_train, ytrain_pred)
mse_train
```
