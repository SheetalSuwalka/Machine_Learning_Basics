```python
import pandas as pd

# Data representing various brand/model structures
data = {
    'Full_Name': [
        'Honda City',           # Standard 2-word
        'Hyundai i10 pro',      # Multi-word model
        'BMW X7',               # Short model
        'Audi Q7',              # Short model
        'Mercedes-Benz C-Class',# Hyphenated brand and model
        'Tesla Model 3 Long Range', # Very long model name
        'xyz-ab model',         # Your custom example
        'Ferrari'               # Edge case: No space at all
    ]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['expand'] = df['Full_Name'].str.split(" ",n=1)
df['brand'] = df['expand'].apply(lambda x: x[0])
df['model'] = df['expand'].apply(lambda x: x[1] if len(x) > 1 else 'NA')

df
```
