```python
k = 'MMMManishhhs'
from collections import Counter

d = Counter(k)

d = dict()

for ch in k:
    if ch in d:
        d[ch]+=1
    else:
        d[ch]=1


# print(d)

sort_dict = dict(sorted(d.items(), key= lambda x:x[1], reverse=True))

s = ""
for k,i in sort_dict.items():
    s += k*i

print(s)
```
