```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import os

from pandas.plotting import autocorrelation_plot
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

```python
os.getcwd()
```

```python
df = pd.read_excel('Retention report Indmoney.xlsx')
```

```python
df.drop('Day', axis = 1)
```

#### E- Sign forecasting using ARIMA Model

```python
# Plot 
plt.figure(figsize=(10, 6))
df.Esign.plot()
plt.title('E-Sign')
plt.xlabel('Day')
plt.ylabel('E-sign Trend')
plt.grid(True)
plt.show()

# Plot ACF
plt.figure(figsize=(10, 6))
plot_acf(df.Esign, lags=15)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()

# Plot PACF
plt.figure(figsize=(10, 6))
plot_pacf(df.Esign, lags=8)
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.grid(True)
plt.show()
### need to decide p (AR), d(I), and q(MA) value
### ACF curve help to decide q value
### PACF curve help to decide p value - 4
```

```python
# Stationarity check of the series
result = adfuller(df.Esign.dropna())
print('p-value: ', result[1])
result = adfuller(df.Esign.diff().dropna())
print('1st order diff p-value: ', result[1])
result = adfuller(df.Esign.diff().diff().dropna())
print('2nd order diff p-value: ', result[1])

### Series is stationary
```

```python
D = 30
df_esign = df.Esign
for i in range(D):
    X = df_esign.values
    size = int(len(X) * 0.85)
    train, test = df_esign[0:size], df_esign[size:]
    # walk-forward validation
    model = ARIMA(df_esign, order=(4,1,2))
    result = model.fit()
    forecast = result.forecast(steps=1)
#     df_esign = df_esign.concate([df_esign,forecast.head(1)], axis = 0)
    df_esign = pd.concat([df_esign,forecast.head(1)], axis = 0)
    df_esign = df_esign.reset_index(drop=True)
    
print(result.summary())
# line plot of residuals
residuals = DataFrame(result.resid)
residuals.plot()
pyplot.show()
# density plot of residuals
residuals.plot(kind='kde')
pyplot.show()
# summary stats of residuals
print(residuals.describe())
```

#### Trades forecasting using ARIMA Model

```python
# Plot 
plt.figure(figsize=(10, 6))
df.Trades.plot()
plt.title('E-Sign Trend')
plt.xlabel('Day')
plt.ylabel('E-sign count')
plt.grid(True)
plt.show()

# Plot ACF
plt.figure(figsize=(10, 6))
plot_acf(df.Trades, lags=15)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()

# Plot PACF
plt.figure(figsize=(10, 6))
plot_pacf(df.Trades, lags=8)
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.grid(True)
plt.show()
### need to decide p (AR), d(I), and q(MA) value
### ACF curve help to decide q value
### PACF curve help to decide p value - 5
```

```python
# Stationarity check of the series
result = adfuller(df.Trades.dropna())
print('p-value: ', result[1])
result = adfuller(df.Trades.diff().dropna())
print('1st order diff p-value: ', result[1])
result = adfuller(df.Trades.diff().diff().dropna())
print('2nd order diff p-value: ', result[1])

### Series is stationary
### First order differencing is stationary, d = 1
```

```python
D = 30
df_trades = df.Trades
for i in range(D):
    X = df_trades.values
    size = int(len(X) * 0.85)
    train, test = df_trades[0:size], df_trades[size:]
    # walk-forward validation
    model = ARIMA(df_trades, order=(5,1,2))
    result = model.fit()
    forecast = result.forecast(steps=1)
#     df_esign = df_esign.concate([df_esign,forecast.head(1)], axis = 0)
    df_trades = pd.concat([df_trades,forecast.head(1)], axis = 0)
    df_trades = df_trades.reset_index(drop=True)
    
print(result.summary())
# line plot of residuals
residuals = DataFrame(result.resid)
residuals.plot()
pyplot.show()
# density plot of residuals
residuals.plot(kind='kde')
pyplot.show()
# summary stats of residuals
print(residuals.describe())
```

```python
# Plot train, test, and forecast
plt.figure(figsize=(10, 6))
plt.plot(df_trades[0:18], label='Training Data')
plt.plot(df_trades[17:], color='red', label='Forecast')
# plt.plot(test.index, forecast, color='red', label='Forecast')
plt.title('Forecasting Trades for future dates')
plt.xlabel('Day')
plt.ylabel('Trade count')
plt.legend()
plt.grid(True)
plt.show()
```
